# Câu 1: Tỉ lệ học sinh tốt nghiệp THPT loại giỏi, khá, trung bình, yếu, rớt ?
 <ul>
    <ul>
    <li> Tính Điểm TN của ba môn bắt buộc Toán Văn Anh và tổ hợp KHTN|KHXH </li>
    <li>Giỏi : Điểm TN >= 8 </li> 
    <li>Khá : 6.5 <= Điểm TN < 8 </li>
    <li> Trung Bình: 5 <= Điểm TN < 6.5 </li>
    <li>Yếu : Điểm TN < 5 </li>
    <li>Rớt : Một trong Điểm Toán, Ngữ văn, Vật lí, Hóa học, Sinh học,Ngoại ngữ,Lịch sử, Địa lí, GDCD < 1.2 </li>
    </ul>
</ul>

```python
KHTN_Points= (df['Toán'] + df['Ngữ văn'] + df['Ngoại ngữ'] + df['KHTN'])/ 4
KHXH_Points= (df['Toán'] + df['Ngữ văn'] + df['Ngoại ngữ'] + df['KHXH'])/ 4
df['Điểm TN']= pd.DataFrame ({"KHTN": KHTN_Points, "KHXH": KHXH_Points}).max(axis= 1)
df.dropna(subset='Điểm TN', inplace= True)
Gioi = df.loc[df['Điểm TN'] >= 8]

Kha = df.loc[(df['Điểm TN'] >= 6.5) & (df['Điểm TN'] <8)]

TrungBinh = df.loc[(df['Điểm TN'] >= 5) & (df['Điểm TN'] < 6.5)]

Yeu = df.loc[df['Điểm TN'] < 5]

Rot= df.loc[(df['Toán'] < 1.2) | (df['Ngữ văn'] < 1.2) | (df['Vật lí'] < 1.2) | (df['Hóa học'] < 1.2) | (df['Sinh học'] < 1.2) | 
        (df['Ngoại ngữ'] < 1.2) | (df['Lịch sử'] < 1.2) | (df['Địa lí'] < 1.2) | (df['GDCD'] < 1.2) ]
        
Ti_Le_Gioi= round(Gioi.value_counts('SBD').sum()/ len(df['Điểm TN']),3)
Ti_Le_Kha= round(Kha.value_counts('SBD').sum()/ len(df['Điểm TN']),3)
Ti_Le_TrungBinh= round(TrungBinh.value_counts('SBD').sum()/ len(df['Điểm TN']),3)
Ti_Le_Yeu= round(Yeu.value_counts('SBD').sum()/ len(df['Điểm TN']), 3)
Ti_le_Rot = round(Rot.value_counts('SBD').sum()/ len(df['Điểm TN']),3)

_Gioi= Gioi.value_counts('SBD').sum()
_Kha= Kha.value_counts('SBD').sum()
_TrungBinh= TrungBinh.value_counts('SBD').sum()
_Yeu= Yeu.value_counts('SBD').sum()
_Rot= Rot.value_counts('SBD').sum()

TiLe= [Ti_Le_Gioi,Ti_Le_Kha,Ti_Le_TrungBinh,Ti_Le_Yeu, Ti_le_Rot]
tmp = [{"Loại": "Giỏi", "Số lượng": _Gioi},
        {"Loại": "Khá", "Số lượng": _Kha},
        {"Loại": "Trung Bình", "Số lượng": _TrungBinh},
        {"Loại": "Yếu", "Số lượng": _Yeu},
        {"Loại": "Rớt TN", "Số lượng": _Rot}]
results_3 = pd.DataFrame.from_dict(tmp).set_index('Loại')
results_3
```

```python
results_3.plot(kind='bar', rot = 60, color ='green')
plt.ylabel('Số lượng thí sinh')
plt.show()
```

        <li>Theo sau là số lượng học sinh tốt nghiệp loại Khá(>300000)  </li>

### Câu 2: Tỉ lệ học sinh thi KHTN so với tỉ lệ học sinh thi KHXH ?
#### Các bước thực hiện: 
- Tính số lượng thí sinh thi KHTN 
- Tính số lượng thí sinh thi KHXH 
- Tính tỉ lệ bằng cách lấy số lượng thí sinh thi KHTN hoặc KHXH chia cho tổng số thí sinh 

```python
khtn = df[['SBD', 'KHTN']]
khtn= khtn.groupby('SBD').value_counts('KHTN').sum()
khxh = df[['SBD', 'KHXH']]
khxh= khxh.groupby('SBD').value_counts('KHXH').sum()
percentKHTN = round(khtn/ len(df),2)
percentKHXH = round(khxh / len(df),2)
percent = [percentKHTN, percentKHXH]
temp = [{'Name': 'KHTN', 'Percent': percentKHTN},
        {'Name': 'KHXH', 'Percent': percentKHXH}]
results = pd.DataFrame.from_dict(temp)
results
```

```python
labels = 'KHTN', 'KHXH'
sizes = percent
plt.figure(figsize=(10,10))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.legend()
plt.show()
```
